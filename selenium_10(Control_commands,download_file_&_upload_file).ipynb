# **Control Commands**

Screenshot 2025-03-11 102358.png

**Note**:The provided code automates copying text from one input field to another on the Text Compare website using Selenium in Python. It starts by initializing the Chrome WebDriver and opening the "https://text-compare.com/" page. After maximizing the window and setting an implicit wait time of 15 seconds to allow the page to load, the script locates the two input fields using their XPath. The first input field is populated with the text "Hello World welcome to my world". Using ActionChains, the script performs a sequence of keyboard actions: it selects all the text with CTRL + A, copies it with CTRL + C, switches to the second input field using the TAB key, and finally pastes the copied text using CTRL + V. The script waits for five seconds to allow observation of the result before closing the browser window. This automation demonstrates how to interact with web elements and simulate keyboard shortcuts using Selenium.



```
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time

# Initialize WebDriver (Ensure chromedriver is properly set up)
driver = webdriver.Chrome()

# Open the W3Schools demo page
driver.get("https://text-compare.com/")
driver.maximize_window()
driver.implicitly_wait(15)

input1=driver.find_element(By.XPATH,"//textarea[@id='inputText1']")
input2=driver.find_element(By.XPATH,"//textarea[@id='inputText2']")
input1.send_keys("Hello World welcome to my world")
act=ActionChains(driver)

# Perform CTRL + A to select all text
act.key_down(Keys.CONTROL).send_keys("a").key_up(Keys.CONTROL).perform()

# perform CTRL + C to copy the text
act.key_down(Keys.CONTROL).send_keys("c").key_up(Keys.CONTROL).perform()

#press tabb key to switch to input2
act.send_keys(Keys.TAB).perform()

#perform CTRL + V to paste the text
act.key_down(Keys.CONTROL).send_keys("v").key_up(Keys.CONTROL).perform()
time.sleep(5)

# Close the browser
driver.quit()

```

# **Download File**

Screenshot 2025-03-11 152013.png



```
import time
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

def chrome_setup():
    location = os.path.abspath(os.getcwd())  # Ensure the absolute path

    # Download files in the desired location
    preferences = {
        "download.default_directory": location,
        "safebrowsing.enabled": True  # Ensure safe downloads
    }

    ops = webdriver.ChromeOptions()
    ops.add_experimental_option("prefs", preferences)

    # Initialize Chrome driver with options
    driver = webdriver.Chrome(options=ops)

    return driver

# Initialize the driver
driver = chrome_setup()

try:
    # Open the target webpage
    driver.get("https://file-examples.com/index.php/sample-documents-download/sample-doc-download/")

    driver.maximize_window()

    # Locate and click the download link
    download_link = driver.find_element(By.XPATH, "//tbody/tr[1]/td[5]/a[1]")
    download_link.click()

    time.sleep(10)  # Wait for the download to complete

except NoSuchElementException:
    print("Download link not found.")

finally:
    driver.quit()  # Ensure browser closes
```
