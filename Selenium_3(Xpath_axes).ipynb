# **what is mean by XPath axes?**

**XPath axes help you navigate through elements in an XML or HTML document relative to another element. Instead of searching the entire page, you can move up, down, or sideways in the DOM (Document Object Model).**

# **Example:**



```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# 1) Open Web Browser (Chrome)
service = Service()  
driver = webdriver.Chrome(service=service)  

# 2) Open the website
driver.get("https://money.rediff.com/gainers")
time.sleep(10)  # Waiting for elements to load (Not recommended, but kept as per your style)

# Maximize window
driver.maximize_window()
time.sleep(10)  # Allow some time after maximizing

# 3) Locate the element and get text by using self
text_msg = driver.find_element(By.XPATH, "//a[contains(text(),'Enterprise Inter')]//self::a").text
print(text_msg)

# 4) Locate the element and get text by using parent
text_msg1 = driver.find_element(By.XPATH, "//a[contains(text(),'Enterprise Inter')]//parent::td").text
print(text_msg1)

# 5) Locate the element and get text by using child
text_msg2 = driver.find_elements(By.XPATH, "//a[contains(text(),'Enterprise Inter')]/ancestor::tr/td")

# 6) Print the number of <td> elements found
print("Total <td> elements:", len(text_msg2))

# 7) Locate the element and get text by using ancestor
row_element = driver.find_element(By.XPATH, "//a[contains(text(),'Enterprise Inter')]/ancestor::tr")

# 6) Get the text content of the entire row
text_msg3 = row_element.text
print("Row Text:", text_msg3)


# 7) Locate the element and get text by using descent
text_msg4 = driver.find_elements(By.XPATH, "//a[contains(text(),'Enterprise Inter')]/ancestor::tr/descendant::*")

# Print the number of descendant elements found
print("Total descendant elements:", len(text_msg4))

# 8) Locate the element and get text by using following
text_msg5 = driver.find_elements(By.XPATH, "//a[contains(text(),'Enterprise Inter')]/ancestor::tr/following::*")

# Print the number of descendant elements found
print("Total following elements:", len(text_msg5))

# 9) Locate the element and get text by using following-sibling
text_msg5 = driver.find_elements(By.XPATH, "//a[contains(text(),'Enterprise Inter')]/ancestor::tr/following-sibling::*")

# Print the number of descendant elements found
print("Total following-sibling elements:", len(text_msg5))

# 10) Locate the element and get text by using preceding
text_msg5 = driver.find_elements(By.XPATH, "//a[contains(text(),'Enterprise Inter')]/ancestor::tr/preceding::*")

# Print the number of descendant elements found
print("Total preceding elements:", len(text_msg5))

# 10) Locate the element and get text by using preceding-sibling
text_msg5 = driver.find_elements(By.XPATH, "//a[contains(text(),'Enterprise Inter')]/ancestor::tr/preceding-sibling::*")

# Print the number of descendant elements found
print("Total preceding-sibling elements:", len(text_msg5))


time.sleep(10)  # Waiting before closing (if needed)
driver.quit()  # Close the browser

```

# **If no child elements have values, you can make the XPath start from ancestor by default.**

**What This Means**

If child elements exist and have text, extract them.

If all child elements are empty, fall back to the ancestor's text.

This ensures data is always available from either child or ancestor.
