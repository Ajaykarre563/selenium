# **what is meann by data picker ?**

**A datepicker in Selenium is a calendar popup on a webpage where you can choose a date (like your birthday or a booking date).**

**In Selenium, you can use code to open the datepicker, navigate the calendar, and select a date automatically.**

# **Example:**

**When booking a flight or hotel online, you often see a calendar pop up to select your travel datesthats a datepicker.**

 **In Selenium, you can interact with a datepicker by:**

Clicking on the input box to open the calendar.

Navigating to the correct month and year.

Selecting the desired date.



```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# 1) Initialize WebDriver
service = Service()  
driver = webdriver.Chrome(service=service)

# 2) Open the website
driver.get("https://jqueryui.com/datepicker/")
driver.implicitly_wait(10)  # Implicit wait to allow elements to load
driver.maximize_window()  # Maximize browser window

# 3) Switch to the correct iframe (fixing the frame switch issue)
iframe = driver.find_element(By.TAG_NAME, "iframe")
driver.switch_to.frame(iframe)

# 4) Locate the datepicker and input the date (fixed XPath syntax error)
#date_input = driver.find_element(By.ID, "datepicker")
#date_input.send_keys("05/21/2022")

# 5) Locate the datepicker and input the date (fixed XPath syntax error)
year="2022"
month="May"
day="21"
driver.find_element(By.ID, "datepicker").click()
# Navigate through the datepicker
while True:
    # Get current month and year
    current_month = driver.find_element(By.XPATH, "//span[@class='ui-datepicker-month']").text
    current_year = driver.find_element(By.XPATH, "//span[@class='ui-datepicker-year']").text

    # Break the loop if month and year match
    if current_month == month and current_year == year:
        break
    else:
        # Click the next button to navigate to the target month/year
        #driver.find_element(By.XPATH, "//*[@id='ui-datepicker-div']/div/a[2]/span").click() # Syntax error for future month
        driver.find_element(By.XPATH, "//*[@id='ui-datepicker-div']/div/a[1]/span").click()  #syntax error for previous month

# Select the day
dates = driver.find_elements(By.XPATH, "//table[@class='ui-datepicker-calendar']//a")
for date in dates:
    if date.text == day:
        date.click()
        break

time.sleep(5)  # Wait to observe the input
driver.quit()  # Cleanly exit the browser
```

# **what is mean by non-standard data picker?**

```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time

# 1) Initialize WebDriver
service = Service()
driver = webdriver.Chrome(service=service)

# 2) Open the website
driver.get("https://www.dummyticket.com/dummy-ticket-for-visa-application/")
driver.implicitly_wait(10)  # Implicit wait to allow elements to load
driver.maximize_window()  # Maximize browser window

# 3) Open the date picker (fixed By.ID to By.XPATH)
driver.find_element(By.XPATH, "//input[@id='dob']").click()

# 4) Select month (December)
datepicker_month = Select(driver.find_element(By.XPATH, "//select[@data-handler='selectMonth']"))
datepicker_month.select_by_visible_text("Dec")

# 5) Select year (1990)
datepicker_year = Select(driver.find_element(By.XPATH, "//select[@data-handler='selectYear']"))
datepicker_year.select_by_visible_text("1990")

# 6) Select day (25)
alldates = driver.find_elements(By.XPATH, "//div[@id='ui-datepicker-div']//table/tbody/tr/td/a")

for date in alldates:
    if date.text == "25":
        date.click()
        break

# 7) Wait and close browser
time.sleep(5)  
driver.quit()

```
