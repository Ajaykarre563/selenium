# **what is mean by Selenium, Locators?**

**In Selenium, Locators are used to find and interact with elements on a web page, like buttons, text fields, or links.**

Screenshot 2025-02-21 183517.png

# **ID Locator**

Explanation: Every HTML element can have a unique id attribute. This is the fastest and most reliable way to find an element.



```
driver.find_element(By.ID, "username").send_keys("Ajay")

```



# **Name Locator**

Explanation: Some elements have a name attribute, which can be used to locate them.



```
driver.find_element(By.NAME, "password").send_keys("mypassword")

```



# **LinkText & PartialLinkText**

Explanation: Used for finding links (a tags).

LinkText matches the full text of a link.

PartialLinkText matches a part of the link text.



```
driver.find_element(By.LINK_TEXT, "Click Here").click()
driver.find_element(By.PARTIAL_LINK_TEXT, "Click").click()

```



# **Example:**



```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# 1) Open Web Browser (Chrome)
service = Service()  
driver = webdriver.Chrome(service=service)  

# 2) Open URL
driver.get("https://demo.nopcommerce.com/")
time.sleep(2)  # Short wait to allow page loading
driver.maximize_window()
time.sleep(2)  # Wait after maximizing

# 3) Search for "Lenovo Thinkpad Carbon Laptop"
driver.find_element(By.NAME, "q").send_keys("Lenovo Thinkpad Carbon Laptop")
time.sleep(2)  # Small delay before clicking

# 4) Click the "Register" link using LINK_TEXT
driver.find_element(By.LINK_TEXT, "Register").click()
time.sleep(2)  # Small delay after clicking

# 5) Click the "Register" link using PARTIAL_LINK_TEXT (Alternative)
# driver.find_element(By.PARTIAL_LINK_TEXT, "Regis").click()
# time.sleep(2)

# Optional: Keep the browser open for a while before quitting
input("Press Enter to close the browser...")

# Close the browser
driver.quit()

```



# **Class Name Locator**

Explanation: Used to locate elements using their class attribute



```
driver.find_element(By.CLASS_NAME, "login-button").click()

```



# **Tag Name Locator**

Explanation: Finds elements by their HTML tag, like input, button, div.



```
driver.find_element(By.TAG_NAME, "button").click()

```



# **Example**



```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# 1) Open Web Browser (Chrome)
service = Service()  
driver = webdriver.Chrome(service=service)  


# 2) Open URL
driver.get("https://www.mystore.in/en/")
time.sleep(10)  # Reduced sleep time to 10 seconds (1000s is too long)

# Maximize window
driver.maximize_window()
time.sleep(5)  # Allow some time after maximizing

# Find elements with class name (Fixed selector)
sliders = driver.find_elements(By.CLASS_NAME, "home-layout")  # Removed "." before class name
print(len(sliders))  # Print the number of sliders
time.sleep(5)  # Short sleep to see output

# Find all links on the page
links = driver.find_elements(By.TAG_NAME, "a")  # Fixed syntax error
print(len(links))
time.sleep(5)

# Close the browser
driver.quit()

```



# **XPath (Most Powerful Locator)**

Explanation: XPath is used to navigate through an HTML page like a tree structure.

**Two Types:**
Absolute XPath – Full path from the root (html).




```
driver.find_element(By.XPATH, "/html/body/div/input").send_keys("Ajay")

```



**Relative XPath** – Shorter, flexible path.



```
driver.find_element(By.XPATH, "//input[@id='username']").send_keys("Ajay")

```



# **What is XPath in Selenium?**

XPath (XML Path) is a way to locate elements on a webpage using their structure in the HTML/XML document. It helps Selenium find elements that may not have unique IDs or class names.

Screenshot 2025-02-22 183810.png



```
driver.find_element(By.XPATH, "//input[@id='username']")

```



**Here’s how it works:**

//input → Finds all input tag elements.

[@id='username'] → Filters to the element where id="username".

# **Types of XPath in Selenium**

**There are two types of XPath:**

# **Absolute XPath (Fixed Path)**

Starts from the root (/) and goes step-by-step to the target element.

**Example:**




```
driver.find_element(By.XPATH, "/html/body/div/input")

```



**Issue: If the structure changes, this XPath breaks.**

# **Relative XPath (Flexible Path) ✅ Recommended**

Starts from anywhere in the HTML using //, making it more flexible.

**Example:**



```
driver.find_element(By.XPATH, "//input[@id='username']")

```



**Benefit: Works even if the structure changes slightly.**
