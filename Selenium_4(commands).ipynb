# **Note:**

Conditional Commands → Work on Web Elements (like buttons, checkboxes, radio buttons).

All Other Commands (Application, Browser, Navigation, Wait) → Are Driver Commands (control the browser).

# **Application Commands in Selenium:**

get(url) → Opens the given URL in the browser.

title → Gets the title of the current web page.

current_url → Returns the URL of the current web page.

page_source → Gets the HTML source code of the web page.



```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# 1) Open Web Browser (Chrome)
service = Service()  
driver = webdriver.Chrome(service=service)  

# 2) Open the website
driver.get("https://opensource-demo.orangehrmlive.com/")
time.sleep(10)  # Waiting for elements to load (Not recommended, but kept as per your style)

# Maximize window
driver.maximize_window()
time.sleep(10)  # Allow some time after maximizing

print(driver.title)
print(driver.current_url)
print(driver.page_source)
```

#**Conditional Commands**

**Used to check conditions on web elements.**

is_displayed() → Checks if an element is visible.

is_enabled() → Checks if an element is enabled (clickable).

is_selected() → Checks if a checkbox, radio button, or dropdown is selected.




```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# 1) Open Web Browser (Chrome)
service = Service()  
driver = webdriver.Chrome(service=service)  

# 2) Open the website
driver.get("https://demo.nopcommerce.com/register")
time.sleep(10)  # Waiting for elements to load (Not recommended, but kept as per your style)

# Maximize window
driver.maximize_window()
time.sleep(10)  # Allow some time after maximizing


#these element through the web element.
search=driver.find_element(By.XPATH,"//input[@id='small-searchterms']")
print("Display status:",search.is_displayed())
print("enable status:",search.is_enabled())

#these is used for checkbox the radio button like male and female
# Locate radio buttons
rb_male = driver.find_element(By.XPATH, "//input[@id='gender-male']")
rb_female = driver.find_element(By.XPATH, "//input[@id='gender-female']")

# Print if they are selected
print(rb_male.is_selected())  
print(rb_female.is_selected())

rb_male.click()
print("After selecting the male radio button:")
print(rb_male.is_selected())  
print(rb_female.is_selected())

rb_female.click()
print("After selecting the female radio button:")
print(rb_male.is_selected())  
print(rb_female.is_selected())


```

# **Browser Commands**

**Used to interact with the browser.**

close() → Closes the current browser window.

quit() → Closes all browser windows and ends the session.



```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# 1) Open Web Browser (Chrome)
service = Service()  
driver = webdriver.Chrome(service=service)  

# 2) Open the website
driver.get("https://opensource-demo.orangehrmlive.com/")
time.sleep(10)  # Waiting for elements to load (Not recommended, but kept as per your style)

# Maximize window
driver.maximize_window()
time.sleep(10)  # Allow some time after maximizing

# Click the link using LINK_TEXT
driver.find_element(By.LINK_TEXT, "OrangeHRM, Inc").click()
time.sleep(10)

# Wait for 10 seconds
time.sleep(10)
# Close the browser
#driver.close() # it close the previous browser the child browser are till open it will close only one browser
#time.sleep(10)

driver.quit() # it will close the multiple browser at a time(this will kill the processor)
```

**Used to navigate through browser history.**

back() → Goes back to the previous page.

forward() → Moves forward to the next page.

refresh() → Reloads the current page.




```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# 1) Open Web Browser (Chrome)
service = Service()  
driver = webdriver.Chrome(service=service)  

# 2) Open the website
driver.get("https://www.snapdeal.com/")
time.sleep(10)

# Maximize window
driver.maximize_window()

time.sleep(10)  # Allow some time after maximizing
driver.get("https://www.amazon.in/")
time.sleep(10)  # Waiting for elements to load (Not recommended, but kept as per your style)

driver.back() #snapdeal page it go
driver.forward() # amazon page it go
driver.refresh()
time.sleep(5)

driver.quit()
```

# **Wait Commands**

**Used to handle page load timing issues.**

implicitly_wait(time) → Waits for all elements to load before throwing an error.

explicit_wait → Waits for a specific element based on a condition.

fluent_wait → Waits for an element with polling intervals and a timeout.

Screenshot 2025-02-24 150057.png

Screenshot 2025-02-24 150038.png

# **Example:**



```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# 1) Open Web Browser (Chrome)
service = Service()
driver = webdriver.Chrome(service=service)

# 2) Open Google
driver.get("https://www.google.co.in/")
driver.implicitly_wait(10) # it will all the statements it will 10 sec for each each wait is 0


# 3) Maximize window
driver.maximize_window()

# 4) Search for "selenium"
search = driver.find_element(By.NAME, 'q')
search.send_keys("selenium")
search.submit()


# 5) Click on Selenium link (More flexible XPath)
driver.find_element(By.XPATH, "//h3[text()='Selenium']").click()

# Optional: Close browser
driver.quit()

```

# **Example:**



```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 1) Open Web Browser (Chrome)
service = Service()
driver = webdriver.Chrome(service=service)

# 2) Open Google
driver.get("https://www.google.co.in/")

# Explicit wait declaration
mywait = WebDriverWait(driver, 10)  # FIXED SPELLING


# 3) Maximize window
driver.maximize_window()

# 4) Search for "selenium"
search = driver.find_element(By.NAME, 'q')
search.send_keys("selenium")
search.submit()

# 5) Wait for search results and click on "Selenium" link
searchlink = mywait.until(EC.element_to_be_clickable((By.XPATH, "//h3[text()='Selenium']")))
searchlink.click()

# Optional: Close browser
driver.quit()

```
