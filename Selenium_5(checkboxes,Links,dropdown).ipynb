# **what is checkboxes in selenium?**

**In Selenium, checkboxes are elements that allow users to select multiple options from a set. You can interact with checkboxes using Selenium WebDriver by selecting or deselecting them.**

How to Handle Checkboxes in Selenium?

**Locate the Checkbox** – Use a locator (like XPath, ID, Name, CSS Selector).

**Select the Checkbox** – Use .click() to check the checkbox.

**Check if it's Selected** – Use .is_selected() to verify its status.

**Deselect (if needed)** – Click again to uncheck (if it's not a fixed selection).



```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# 1) Open Web Browser (Chrome)
service = Service()
driver = webdriver.Chrome(service=service)

# 2) Open the website
driver.get("https://testautomationpractice.blogspot.com/")
driver.implicitly_wait(10)

# 3) Maximize window
driver.maximize_window()

# 4) Select a specific checkbox
driver.find_element(By.XPATH, "//label[@for='monday']").click()

# Optional: Wait a few seconds to observe the result
#time.sleep(5)

# 5) Select all checkboxes
checkboxes = driver.find_elements(By.XPATH, "//input[@type='checkbox' and contains(@id,'day')]")
print(len(checkboxes))  # Print the number of checkboxes found

# Click each checkbox
for checkbox in checkboxes:
    checkbox.click()

# 6)select multiple checkbox by choice
#for checkbox in checkboxes:
 #   weekname=checkbox.get_attribute('id')
  #  if weekname == 'monday' or weekname=='sunday':
   #     checkbox.click()

# 7)select the last two checkboxes
#range(5,7)-->(6,7)
#for i in range(len(checkboxes)-2,len(checkboxes)):
 #   checkboxes[i].click()

# 8)select first two checkboxes
#for i in range(len(checkboxes)):
 #   if i<2:
  #      checkboxes[i].click()
  
time.sleep(5)
# 9)clear all the checkboxes
for checkbox in checkboxes:
    if checkbox.is_selected():
        checkbox.click()




# 5) Close the browser
driver.quit()

```

A link that directs users to a different website.

Example: A blog linking to an external news site.



```
from selenium import webdriver  # Import Selenium WebDriver
from selenium.webdriver.chrome.service import Service  # Import Service to manage WebDriver
from selenium.webdriver.common.by import By  # Import By to locate elements
import time

# 1) Initialize WebDriver (Make sure you have the correct ChromeDriver installed)
service = Service()  # Creates a service instance for ChromeDriver
driver = webdriver.Chrome(service=service)  # Launches Chrome browser

# 2) Open the website
driver.get("https://demo.nopcommerce.com/")  # Navigate to the target website
driver.implicitly_wait(10)  # Implicit wait to allow elements to load

# 3) Maximize window
driver.maximize_window()  # Ensures the browser opens in full screen

# 4) Click on the "Digital downloads" link using PARTIAL_LINK_TEXT locator
driver.find_element(By.PARTIAL_LINK_TEXT, "Digital").click()
#driver.find_element(By.LINK_TEXT, "Digital Downloads").click()

# 5) Find all links on the page using XPath ('//a' selects all anchor tags)
links = driver.find_elements(By.XPATH, '//a')
#links = driver.find_elements(By.LINK_TEXT, 'a')

# 6) Print the total number of links found on the page
print("Total number of links:", len(links))

# 7) Print the text of each link (only non-empty links will be printed)
print("\nList of links on the page:")
time.sleep(10)
for link in links:
    print(link.text)
    

# 8) Close the browser window
driver.close()  # Closes the current browser window

```

A link that points to a page that no longer exists or is incorrect.

Example: A website links to a removed or misspelled



```
<a href="https://example.com/old-page">Old Page</a>

```





```
import requests  # Correct import alias
from selenium import webdriver  # Import Selenium WebDriver
from selenium.webdriver.chrome.service import Service  # Import Service to manage WebDriver
from selenium.webdriver.common.by import By  # Import By to locate elements

# 1) Initialize WebDriver
service = Service()  # Creates a service instance for ChromeDriver
driver = webdriver.Chrome(service=service)  # Launches Chrome browser

# 2) Open the website
driver.get("http://www.deadlinkcity.com/")  # Navigate to the target website
driver.implicitly_wait(10)  # Implicit wait to allow elements to load

# 3) Find all links
all_links = driver.find_elements(By.TAG_NAME, 'a')  # Use find_elements() to get all links
count = 0

for link in all_links:
    url = link.get_attribute('href')  # Get the href attribute

    if not url:  # Skip links with no href
        print("Skipping empty link")
        continue

    try:
        res = requests.head(url, allow_redirects=True)  # Send a HEAD request
        if res.status_code >= 400:
            print(url, "is a broken link")
            count += 1
        else:
            print(url, "is a valid link")
    except requests.RequestException as e:  # Catch request errors
        print(url, "could not be checked due to an error:", str(e))

print("Total number of broken links:", count)

# Close the browser
driver.quit()

```

# **What is a Dropdown in Selenium?**

A dropdown in Selenium refers to a select Tag HTML element that allows users to choose an option from a list. Selenium provides the Select class to interact with dropdowns easily.

**Why Use the Select Class?**

The Select class helps in selecting, deselecting, and retrieving options from a dropdown.

# **Key Takeaways:**

select_by_visible_text("India") → Selects the option with text "India".

select_by_value("10") → Selects the option where value="10".

select_by_index(13) → Selects the 14th option (index starts at 0).



```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time

# 1) Initialize WebDriver
service = Service()  # Creates a service instance for ChromeDriver
driver = webdriver.Chrome(service=service)  # Launches Chrome browser

# 2) Open the website
driver.get("https://www.opencart.com/index.php?route=account/register")  # Fixed URL
driver.implicitly_wait(10)  # Implicit wait

# 3) Maximize window
driver.maximize_window()

# Locate the dropdown element
dropdown = Select(driver.find_element(By.XPATH, "//select[@id='input-country']"))

# Select options from the dropdown
dropdown.select_by_visible_text("India")  # Select by visible text
time.sleep(2)  # Wait to see selection

# Capture all dropdown options and print them
all_options = dropdown.options  # List of options

print(f"Total options: {len(all_options)}")  # Print total count

for opt in all_options:
    print(opt.text)  # Print option text

# select option from dropdown without using the bulit in functions

for opt in all_options:
    if opt.text=="india":
        opt.click()
        break

# Close the driver
driver.quit()

```
