# **what is mean by Software Testing?**

**Software testing is the process of checking if a software application works correctly and has no errors (bugs). It ensures the software is reliable, secure, and meets user needs.**

# **Example**

Imagine you're developing a food delivery app. Testing will check:

✅ Can users sign up and log in successfully?

✅ Does the app show the correct restaurant list?

✅ When a user orders food, does the payment work properly?

✅ Does the app crash if many users use it at the same time?

Testing helps find and fix issues before releasing the app to users! ✅

test.png

# 1️⃣ **Manual Testing**

Testing done by humans without automation tools.

✅ Example: A tester manually checks if a login page works by entering different usernames and passwords.

# 2️⃣ **Automation Testing**

Testing done using tools like Selenium to save time.

✅ Example: A script automatically tests if adding items to a shopping cart works.

3️⃣ **White Box Testing**

Testing based on the internal code of the software.

✅ Example: A developer checks if a function in the code correctly calculates discounts.

4️⃣ **Black Box Testing**

Testing based on software functionality, without looking at the code.

✅ Example: A tester checks if clicking the “Forgot Password” button sends a reset email.

5️⃣ **Grey Box Testing**

A mix of White Box & Black Box testing, where the tester knows some internal code details.

✅ Example: A tester verifies a database update happens correctly when a user updates their profile.

# **Functional Testing (Checks if features work correctly)**

6️⃣ **Unit Testing**

Testing individual components of software.

✅ Example: Checking if a calculator app correctly adds two numbers.

7️⃣ **Integration Testing**

Testing if different parts of the software work together.

✅ Example: Checking if the payment system properly connects with the shopping cart.

🔹 Incremental Testing – Tests parts step by step.

🔹 Non-Incremental Testing – Tests everything at once.

🔹 Top-Down Testing – Tests major parts first, then small parts.

🔹 Bottom-Up Testing – Tests small parts first, then major ones.

8️⃣ **System Testing**

Testing the entire application to check if it works correctly.

✅ Example: Verifying if a mobile banking app allows login, money transfers, and logout properly.

# **Non-Functional Testing (Checks quality aspects)**

9️⃣ **Performance Testing**

Tests speed and stability under different loads.

✅ Example: Checking if a website loads fast when 1,000 users visit at once.

🔹 Load Testing – Tests if the system handles a high number of users.

🔹 Stress Testing – Tests how the system behaves under extreme conditions.

🔹 Scalability Testing – Checks if the system can handle growing users.

🔹 Stability Testing – Checks if the system runs smoothly over time.

🔟 **Usability Testing**

Checks if the software is easy to use.

✅ Example: Testing if an e-learning app has clear navigation for students.

1️⃣1️⃣ **Compatibility Testing**

Tests if the software works on different devices and browsers.

✅ Example: Ensuring a website works on Chrome, Firefox, and Safari.

Final Thought

These different types of testing help ensure software is bug-free, reliable, and user-friendly before launch! ✅

# **what is  mean by Selenium?**

**Selenium is a tool that automates website testing. It helps a computer do what a human would do on a website, like clicking buttons, typing in forms, and checking if things work properly.**

# **Example**

Imagine you have a login page. Instead of testing it manually every time, Selenium can:

Open the browser

Type the username and password

Click the "Login" button

Check if the login was successful

This saves time and ensures the website works correctly every time!

# **what is mean by WebDriver in Selenium?**

**WebDriver is a component in Selenium, and it is also available as a module in Python.**

**WebDriver is a tool in Selenium that helps automate web browsers. It allows scripts to control a browser just like a real user—clicking buttons, entering text, and navigating pages.**

Selenium WebDriver is a core component of Selenium that automates browsers like Chrome, Firefox, and Edge. It interacts with web elements (buttons, forms, links, etc.).

🔹 Example: WebDriver is a key part of Selenium, just like an "engine" in a car.



```
from selenium import webdriver  

# Open Chrome browser
driver = webdriver.Chrome()  

# Go to Google
driver.get("https://www.google.com")  

# Close the browser
driver.quit()

```

1️⃣ **Presentation Layer (GUI/UI)**

This is what the user sees and interacts with (buttons, forms, screens).
Users send requests by clicking or entering data.

Testing Type: ✅ GUI Testing (checks if buttons, forms, and navigation work properly).

✅ **Example:**

A user logs in to an online shopping website by entering a username and password in a login form.

2️⃣ **Application/Web Server (Business Logic Layer)**

This is where the main logic of the application happens.
The server processes the request and sends responses using APIs (Application Programming Interfaces).

Testing Type: ✅ API Testing (checks if data is sent and received correctly between the frontend and backend).

✅ **Example:**

When a user clicks "Login", the application server checks the username and password using an API and returns a response (success or failure).

3️⃣ **Database Layer (Backend)**

This is where all data is stored (user details, orders, payments, etc.).
The application sends queries to the database to store or fetch data.

Testing Type: ✅ Database Testing (checks if data is stored and retrieved correctly).

✅ **Example:**

After a successful login, the user's profile information is fetched from the database and displayed on the screen.

# **Modules:**

selenium.webdriver → Provides the WebDriver to control the browser.

selenium.webdriver.chrome.service.Service → Manages the ChromeDriver service.

selenium.webdriver.common.by.By → Helps locate elements on the webpage.

time → Provides sleep functionality to pause execution.

# **Methods:**

Service() → Initializes the WebDriver service for Chrome.

webdriver.Chrome(service=service) → Launches the Chrome browser.

driver.get(url) → Opens the specified URL in the browser.

driver.find_element(By.NAME, "username").send_keys("Admin") → Finds the username input field and enters text.

driver.find_element(By.NAME, "password").send_keys("admin123") → Finds the password input field and enters text.

driver.find_element(By.XPATH, "//button[@type='submit']").click() → Finds and clicks the login button.

driver.title → Retrieves the title of the current page.

time.sleep(5) → Pauses execution for 5 seconds.

driver.quit() → Closes the browser and ends the WebDriver session.

# **Example:**



```
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# 1) Open Web Browser (Chrome)
service = Service()  
driver = webdriver.Chrome(service=service)  

# 2) Open URL
driver.get("https://opensource-demo.orangehrmlive.com/")
time.sleep(50000)  # Wait for page to load

# 3) Enter username
driver.find_element(By.NAME, "username").send_keys("Admin")
time.sleep(20000)

# 4) Enter password
driver.find_element(By.NAME, "password").send_keys("admin123")
time.sleep(200000)

# 5) Click on Login button
driver.find_element(By.XPATH, "//button[@type='submit']").click()
time.sleep(500000)  # Wait for login process

# 6) Capture title of the home page
act_title = driver.title

# 7) Verify title of the page
exp_title = "OrangeHRM"

if act_title == exp_title:
    print("Login success")
    time.sleep(20000)
else:
    print("Login fail")

# 8) Hold the window for 53 minutes (3180 seconds)
print("Window will stay open for 53 minutes...")
time.sleep(3)

# 9) Close browser
driver.close()

```
