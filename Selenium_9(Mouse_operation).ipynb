# **what is mean by mouse operation?**

**Mouse operations refer to actions you perform using a computer mouse. These actions allow you to interact with items on a screen.**

**Common mouse operations include:**

Click – Pressing and releasing the left mouse button.

Right-click – Pressing and releasing the right mouse button (opens context menus).

Double-click – Quickly clicking the left mouse button twice.

Mouse hover – Moving the pointer over an item without clicking.

Drag and drop – Clicking and holding an item, moving it to a new location, and releasing the button.

Scroll – Moving the scroll wheel to navigate through pages.

# **Mouse Hover**

Screenshot 2025-03-08 131644.png



```
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains

# Initialize WebDriver
driver = webdriver.Chrome()
driver.get("https://opensource-demo.orangehrmlive.com/")
driver.maximize_window()

# Set implicit wait
driver.implicitly_wait(15)

# Locate and interact with login elements
username = driver.find_element(By.NAME, "username")
password = driver.find_element(By.NAME, "password")

# Enter credentials
username.send_keys("Admin")
password.send_keys("admin123")

# Click Login
driver.find_element(By.XPATH, "//button[@type='submit']").click()

# Navigate to Admin -> User Management -> Users
admin = driver.find_element(By.XPATH, "//*[@id='menu_admin_viewAdminModule']/b")
usermgmt = driver.find_element(By.XPATH, "//*[@id='menu_admin_UserManagement']")
users = driver.find_element(By.XPATH, "//*[@id='menu_admin_viewSystemUsers']")

# Perform Mouse Hover
actions = ActionChains(driver) # Initialize ActionChains it is used to perform mouse hover
actions.move_to_element(admin).perform()  # Hover over "Admin"
actions.move_to_element(usermgmt).perform()  # Hover over "User Management"
actions.move_to_element(users).click().perform()  # Click on "Users"

print("Successfully navigated to Users page!")

# Close browser
driver.quit()

```

Screenshot 2025-03-08 163015.png



```
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

# Initialize WebDriver (Ensure chromedriver is properly set up)
driver = webdriver.Chrome()

# Open the W3Schools demo page
driver.get("https://www.w3schools.com/tags/tryit.asp?filename=tryhtml5_ev_ondblclick3")
driver.maximize_window()
driver.implicitly_wait(15)

# Switch to the iframe where the input field and button are present
driver.switch_to.frame("iframeResult")

# Locate the input field, clear it, and enter new text
field1 = driver.find_element(By.XPATH, "//input[@id='field1']")
field1.clear()
field1.send_keys("Ajay Karre")

# Locate the button for double-click action (Improved XPath for accuracy)
button = driver.find_element(By.XPATH, "//button[normalize-space()='Copy Text']")

# Perform double-click action on the button
actions = ActionChains(driver)
actions.double_click(button).perform()

# Wait to observe the result
time.sleep(5)

print("Double click performed successfully.")

# Close the browser
driver.quit()

```

```
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

# Initialize WebDriver (Ensure chromedriver is properly set up)
driver = webdriver.Chrome()

# Open the W3Schools demo page
driver.get("https://www.jqueryscript.net/demo/Price-Range-Slider-jQuery-UI/")
driver.maximize_window()
driver.implicitly_wait(15)

min_slider=driver.find_element(By.XPATH, "//div[@id='slider-range']//span[1]")
max_slider=driver.find_element(By.XPATH, "//div[@id='slider-range']//span[2]")
actions = ActionChains(driver)

print("Initial position of min_slider: ", min_slider.location)
print("Initial position of max_slider: ", max_slider.location)

actions.drag_and_drop_by_offset(min_slider, 100, 0).perform() # Drag min_slider to the right
actions.drag_and_drop_by_offset(max_slider, -40, 0).perform() # Drag max_slider to the left
print("Final position of min_slider: ", min_slider.location)
print("Final position of max_slider: ", max_slider.location)

time.sleep(5)

```

# **scrolling is not mouse operation it can be excuted my javascript.**

Screenshot 2025-03-08 201648.png



```
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Initialize WebDriver (Ensure chromedriver is properly set up)
driver = webdriver.Chrome()

# Open the Flags of the World page
driver.get("https://www.countries-ofthe-world.com/flags-of-the-world.html")
driver.maximize_window()
driver.implicitly_wait(15)

# 1. Scroll down the page by pixel (3000 pixels)
driver.execute_script("window.scrollBy(0,3000);")
time.sleep(2)  # Allow time for scroll action to complete
value = driver.execute_script("return window.pageYOffset;")
print("Number of pixels moved:", value)  # Expected output: 3000

# 2. Scroll down the page until the element is visible (Example: Flag of India)
flag = driver.find_element(By.XPATH, "//img[@alt='Flag of India']")
driver.execute_script("arguments[0].scrollIntoView();", flag)
time.sleep(2)  # Allow time for element to be scrolled into view
value = driver.execute_script("return window.pageYOffset;")
print("Number of pixels moved (Flag of India):", value)

# 3. Scroll down the page to the end
driver.execute_script("window.scrollBy(0, document.body.scrollHeight);")
time.sleep(2)  # Allow time for scroll action to complete
value = driver.execute_script("return window.pageYOffset;")
print("Number of pixels moved (End of page):", value)

# 4. Scroll back up to the starting position
driver.execute_script("window.scrollBy(0, -document.body.scrollHeight);")
time.sleep(2)  # Allow time to return to the top
value = driver.execute_script("return window.pageYOffset;")
print("Back to starting position (pixels moved):", value)  # Expected output: 0

# Close the browser
time.sleep(5)
driver.quit()

```
