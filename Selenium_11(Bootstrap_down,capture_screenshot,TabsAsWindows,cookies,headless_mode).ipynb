# **what is mean by bootstrap down?**

A Bootstrap dropdown is a type of menu that appears when you click a button. It is created using Bootstrap, a popular website styling tool.

In Selenium, handling a Bootstrap dropdown is different because it is made with div and a tags, not the usual select and option tags.

âœ… Simple Example:

**Imagine a dropdown with options like this:**


India

USA

UK

**Steps to select "India":**

Click the dropdown to open it.
Find the option you want.
Click on the option.

```
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Initialize the WebDriver
driver = webdriver.Chrome()
driver.implicitly_wait(10)

# Open the target webpage
driver.get("https://www.dummyticket.com/dummy-ticket-for-visa-application/")
driver.maximize_window()

# Click the dropdown to display country options
driver.find_element(By.XPATH, "//span[@id='select2-billing_country-container']").click()

# Get the list of country options
countries_list = driver.find_elements(By.XPATH, "//ul[@id='select2-billing_country-results']/li")
print(len(countries_list))

# Iterate through the country list and select "India"
for country in countries_list:
    if country.text == "Iran":
        country.click()
        break
time.sleep(5)
# Optional: Close the browser
driver.quit()

```

# **TabsASWindows**



```
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Initialize the WebDriver
driver = webdriver.Chrome()
driver.implicitly_wait(10)

# Open the target webpage
driver.get("https://demo.nopcommerce.com/")
driver.maximize_window()

# Open the "Register" page in a new tab using CTRL + click
reg_link = driver.find_element(By.LINK_TEXT, "Register")
reg_link.send_keys(Keys.CONTROL + Keys.RETURN)
time.sleep(5)

# Switch to the new tab
driver.switch_to.window(driver.window_handles[1])
time.sleep(5)
```

# **what is mean by cookie in selenium?**

**In Selenium, cookies are small pieces of data stored by the browser to remember information about a userâ€™s session.**

ðŸ“Œ **Simple Understanding:**

Cookies help websites remember:

Login Information (e.g., staying logged in).

User Preferences (e.g., language settings).

Session Data (e.g., items in a shopping cart).

**In Selenium, you can:**

Capture Cookies â€“ See what information the website stores.

Add Cookies â€“ Inject your own data for testing.

Delete Cookies â€“ Clear user data for a fresh session.

**âœ… Example:**

When you log in to a website, a cookie is created to keep you logged in. Selenium can:

Get Cookies â€“ Check if the login cookie exists.

Add Cookies â€“ Simulate a logged-in user without manual login.

Delete Cookies â€“ Log out by removing session cookies.



```
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Initialize the WebDriver
driver = webdriver.Chrome()
driver.implicitly_wait(10)

# Open the target webpage
driver.get("https://demo.nopcommerce.com/")
driver.maximize_window()

# Capture cookies from the browser
cookies = driver.get_cookies()

# Print the total number of cookies
print(f"Total Cookies: {len(cookies)}")

# Print cookie details
#for c in cookies:
 #   print(f"Cookie Name: {c.get('name')}")
    #print(f"Cookie Value: {c.get('value')}")
    #print(c)  # Print the entire cookie object (optional for debugging)


# Optional: Add a new cookie
driver.add_cookie({"name": "my_custom_cookie", "value": "test123"})
print("New cookie added successfully!")

# Verify the updated cookies list
updated_cookies = driver.get_cookies()
print(f"Updated Cookies: {len(updated_cookies)}")

#delete a cookie from the browser
driver.delete_cookie("my_custom_cookie")
print("Cookie deleted successfully!")

# Verify the deleted cookie
updated_cookies = driver.get_cookies()
print(f"Updated Cookies: {len(updated_cookies)}")

#delete all cookie
#driver.delete_all_cookies() #delete all cookies
#print("All cookies deleted successfully!")

# Close the browser
driver.quit()

```

# **What is "Headless" Mode in Selenium?**

**Headless mode in Selenium means running a browser without a visible user interface (UI). The browser runs in the background, performing actions like opening pages, clicking buttons, and scraping data without displaying a window.**

**Why Use Headless Mode?**

âœ… Faster Execution â†’ No need to render graphics/UI.

âœ… Automated Testing â†’ Useful for running tests on servers without a display.

âœ… Web Scraping â†’ Avoids pop-ups, distractions, and runs efficiently.

âœ… CI/CD Pipelines â†’ Ideal for Jenkins, GitHub Actions, or Docker environments.



# **It will not show the browser running.. it will directly perform operations**



```
from selenium import webdriver

# Function to launch Chrome in headless mode
def headless_chrome():
    from selenium.webdriver.chrome.service import Service
    serv_obj = Service()
    # Set Chrome options for headless mode
    ops = webdriver.ChromeOptions()
    ops.headless = True  # Enables headless mode

    # Initialize WebDriver with Service and Options
    driver = webdriver.Chrome(service=serv_obj, options=ops)
    return driver

# Launch the browser in headless mode
driver = headless_chrome()

# Open the target webpage
driver.get("https://demo.nopcommerce.com/")
# Output page title and URL
print("Page Title: ", driver.title)
print("Current URL: ", driver.current_url)

# Close the browser
driver.quit()

```
